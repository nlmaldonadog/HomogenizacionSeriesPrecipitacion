# Incrementos anuales
Cuaderno de texto para abordar la generacion de las graficas de incrementos anuales, para poder ejecutar este cuaderno es necesario primero ejecutar el cuaderno de Modelos SARIMA

```python
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.stattools import pacf
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_pacf
import matplotlib.pyplot as plt
from pmdarima import auto_arima
import shutil
import os
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')

lista_zonas = ['MAGDALENA_MEDIO']
path = '../datos/imputados_SARIMA/'
```

```python
for zona in lista_zonas:
    print('Zona:',zona)
    path_zone = path+zona

    shutil.rmtree(f'../datos/incrementos_anuales/{zona}/', ignore_errors=True)
    path_output = f'../datos/incrementos_anuales/{zona}/'
    os.makedirs(path_output)

    for file in os.listdir(path_zone):
        print('Estacion:',file)
        df = pd.read_csv(path_zone+'/'+file)

        estacion = file[:-4]

        df['Fecha'] = pd.to_datetime(df['Fecha'])
        df.set_index('Fecha', inplace=True)

        # Calcular los incrementos anuales
        df['Incrementos_Anuales'] = df['Precipitaci贸n_imputada'].diff(12)
        
        image_path = f'../salidas/{zona}/{estacion}/SARIMA/incrementos_anuales.jpg'
        if os.path.exists(image_path):
            os.remove(image_path)

        fig, ax = plt.subplots(2, 1, figsize=(12, 8))
        ax[0].plot(df['Precipitaci贸n_imputada'], label='Precipitaci贸n imputada')
        ax[0].set_title('Precipitaci贸n imputada')
        ax[0].legend()
        ax[1].plot(df['Incrementos_Anuales'], label='Incrementos anuales')
        ax[1].set_title('Incrementos anuales')
        ax[1].legend()
        plt.tight_layout()
        plt.savefig(image_path)
        plt.close()

        df.to_csv(path_output+file, index=True)
```
