# Identificación de datos faltantes
## y sustitución de los mismos
### Preprocesamiento de los datos
En este cuaderno se va a explorar las formas de preprocesar los datos para imputar los valores nulos.

```python
import shutil
import os
import pandas as pd
import warnings
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

warnings.filterwarnings('ignore')
```

```python
lista_zonas = ['MAGDALENA_MEDIO']
path = '../datos/separados_por_estacion/'
```

```python
for zona in lista_zonas:
    print('Zona:',zona)
    path_zone = path+zona

    shutil.rmtree(f'../datos/imputados/{zona}/', ignore_errors=True)
    path_output = f'../datos/imputados/{zona}/'
    os.makedirs(path_output)

    for file in os.listdir(path_zone):
        print('Archivo:',file)
        col = file.split('.')[0]

        file_path = path_zone + '/' + file
        df = pd.read_csv(file_path, sep=',')

        print(df.head(5))

        # Buscar columnas con valores nulos
        col_null = df.columns[df.isnull().any()].tolist()
        print('Columnas con valores nulos:',col_null)

        # Reemplazar valores nulos por -99999
        df.fillna(-99999, inplace=True)

        # Guardar archivo
        df.to_csv(path_output+col+'.csv', index=False)
```
