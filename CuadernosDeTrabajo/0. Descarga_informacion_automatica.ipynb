```python
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
from sodapy import Socrata

cliente = Socrata("www.datos.gov.co", None)

results = cliente.get("s54a-sgyg", limit=5)
results
```

Exploramos opciones de busqueda sobre el API directamente gracias a la documentación de la libreria [Socrata](https://dev.socrata.com/docs/queries/).

```python
zonas_hidrograficas = ['MEDIO MAGDALENA', 'BAJO MAGDALENA', 'CARIBE - LITORAL', 'SINÚ']
where_condition = f"zonahidrografica IN ({','.join([f'\'{zona}\'' for zona in zonas_hidrograficas])})"
results = cliente.get("s54a-sgyg", where=where_condition, limit=10000)
results[0:5]
```

```python
start_date = '2000-01-01T00:00:00.000'
end_date = '2001-01-31T23:59:59.999'
where_condition = f"fechaobservacion >= '{start_date}' AND fechaobservacion <= '{end_date}' AND zonahidrografica IN ({','.join([f'\'{zona}\'' for zona in zonas_hidrograficas])})"
results = cliente.get("s54a-sgyg", where=where_condition, limit=10000)
results[0:5]
```

Observamso que no se encuenta información del 2000 en nuestra fuente de datos, asi que buscamos el valor mas antiguo presente.

```python
results = cliente.get("s54a-sgyg", order="fechaobservacion", limit=5)
results
```

Encontramos que no cuenta con información anterior al 2003. De igual forma tomamos la información del punto de interes, en este caso las zonas hidrograficas de MEDIO MAGDALENA, BAJO MAGDALENA, CARIBE - LITORAL y SINÚ.

```python
start_date = '2010-01-01T00:00:00.000'
end_date = '2011-01-31T23:59:59.999'
zonas_hidrograficas = ['BAJO MAGDALENA', 'CARIBE - LITORAL', 'SINÚ']
zonas = ','.join([f'\'{zona}\'' for zona in zonas_hidrograficas])
print(zonas)
where_condition = f"fechaobservacion >= '{start_date}' AND fechaobservacion <= '{end_date}' AND zonahidrografica IN ({zonas})"
results = cliente.get("s54a-sgyg", where=where_condition, limit=100000)
results[0:5]
```
